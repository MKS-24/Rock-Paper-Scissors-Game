```python
import streamlit as st
import random

# --- Page Setup ---
st.set_page_config(page_title="Rock Paper Scissors", layout="centered")
st.title("ğŸ® Rock Paper Scissors Game")

# --- Initialize Session State ---
if "game_started" not in st.session_state:
    st.session_state.game_started = False
if "game_over" not in st.session_state:
    st.session_state.game_over = False
if "player_choice" not in st.session_state:
    st.session_state.player_choice = None
if "computer_choice" not in st.session_state:
    st.session_state.computer_choice = None
if "wins" not in st.session_state:
    st.session_state.wins = 0
if "losses" not in st.session_state:
    st.session_state.losses = 0
if "draws" not in st.session_state:
    st.session_state.draws = 0

# --- Ask to Start the Game ---
if not st.session_state.game_started:
    choice = st.radio("Do you want to play the game?", ["Yes", "No"])
    if st.button("Confirm"):
        if choice == "Yes":
            st.session_state.game_started = True
        else:
            st.success("ğŸ‘‹ Allah Hafiz")
            st.stop()

# --- Main Game ---
if st.session_state.game_started and not st.session_state.game_over:
    st.write("### Choose one option below ğŸ‘‡")

    tup = ("Rock", "Paper", "Scissors")
    cols = st.columns(3)
    for i, val in enumerate(tup):
        with cols[i]:
            if st.button(val):
                st.session_state.player_choice = val
                st.session_state.computer_choice = random.choice(tup)

                player = st.session_state.player_choice
                computer = st.session_state.computer_choice

                st.subheader("ğŸ² Results")
                st.info(f"ğŸ–¥ï¸ Computer: **{computer}**")
                st.info(f"ğŸ™‹ You: **{player}**")

                # --- Game Logic ---
                if (
                    (player.lower() == "paper" and computer.lower() == "rock")
                    or (player.lower() == "scissors" and computer.lower() == "paper")
                    or (player.lower() == "rock" and computer.lower() == "scissors")
                ):
                    st.success("ğŸ† You Won!")
                    st.session_state.wins += 1
                elif player.lower() == computer.lower():
                    st.warning("ğŸ¤ It's a Draw!")
                    st.session_state.draws += 1
                else:
                    st.error("ğŸ˜¢ You Lost!")
                    st.session_state.losses += 1

                st.session_state.game_over = True

# --- Scoreboard Section ---
if st.session_state.game_started:
    st.markdown("### ğŸ§¾ Scoreboard")
    col1, col2, col3 = st.columns(3)
    col1.metric("ğŸ† Wins", st.session_state.wins)
    col2.metric("ğŸ˜¢ Losses", st.session_state.losses)
    col3.metric("ğŸ¤ Draws", st.session_state.draws)

# --- Play Again or Exit ---
if st.session_state.game_over:
    col1, col2 = st.columns(2)
    with col1:
        if st.button("ğŸ” Play Again"):
            st.session_state.game_over = False
            st.session_state.player_choice = None
            st.session_state.computer_choice = None
    with col2:
        if st.button("ğŸšª Exit"):
            st.success("ğŸ‘‹ Allah Hafiz")
            st.session_state.game_started = False
```
